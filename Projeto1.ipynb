passo 1: Percorrer todos os arquivos da pasta base de dados (pasta vendas)

```python
import os
import pandas as pd


lista_arquivo = os.listdir("/content/drive/MyDrive/curso python/projeto1/Vendas")
display(lista_arquivo)

```

passo 2: Importar as bases de dados de vendas
passo 3: Tratar as bases de dados / compilar as bases



```python
tabela_total = pd.DataFrame()

for arquivo in lista_arquivo:
  if "Vendas" in arquivo:
    tabela = pd.read_csv(f"/content/drive/MyDrive/curso python/projeto1/Vendas/{arquivo}")
    tabela_total = tabela_total._append(tabela)

display(tabela_total)
```

passo 4: Calcular o produto mais vendido (em quantidade)

```python
tabela_produtos = tabela_total.groupby("Produto").sum()
#grupby agrupa uma coluna
tabela_produtos = tabela_produtos[["Quantidade Vendida"]].sort_values(by="Quantidade Vendida", ascending=False)
#escolhi qual coluna eu queria que aparecesse. Se for uma coluna só, pode usar só um [] (porém vai ficar em um formato diferente), mas se for mais de uma precisa usar dois [[]].
#o .sort_values() ordena os valores, em "by='Quantidade Vendida'", vc escolhe qual coluna dita a ordem e em "ascending=False" vc diz se a ordem é crescente
display(tabela_produtos)
```

passo 5: Calcular o produto que mais faturou (em faturamento)

```python
tabela_total["Faturamento"] = tabela_total['Quantidade Vendida'] * tabela_total['Preco Unitario']
#criei mais uma coluna na planilha, com os valores da quantidade vendida * preço unitário
# "" '' em texto tanto faz usar aspas duplas ou simples
display(tabela_total)
```

```python
tabela_faturamento = tabela_total.groupby("Produto").sum()
#grupby agrupa uma coluna
tabela_faturamento = tabela_faturamento [["Faturamento"]].sort_values(by="Faturamento", ascending=False)
display(tabela_faturamento)
```

passo 6: Caclcular a loja/cidade que mais vendeu - criar grafico

```python
tabela_lojas = tabela_total.groupby("Loja").sum()
#o groupby torna a coluna indicada o indice da tabela
tabela_lojas = tabela_lojas [["Faturamento"]].sort_values(by="Faturamento", ascending=False)
display(tabela_lojas)
```

```python
import plotly.express as px
#é uma bibliotecas para criar gráfico

grafico = px.bar(tabela_lojas, x= tabela_lojas.index, y='Faturamento')
#criou o grafico em barras
#o tabela_lojas.index é para que o px entenda que é pra pegar o indice, pq o pd transforma a coluna indicada em index, como já citado.
#Se nao botar o .index o px não entende que a coluna loja faz parte da tabela
grafico.show()
#pra mostrar o gráfico
```
